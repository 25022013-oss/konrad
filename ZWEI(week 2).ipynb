```python
#W2A15
n=int(input("Nhập n :"))
print("số * tương ứng : ", 6*n*(n-1)+1,end="\n\n")
k=6 + (n-2)*6
l=0
o=2*(n-1)-2
m=6+(n-2)*4
if n == 1 :
  print("*")
elif n > 1:
  for i in range(1,n,1):
   print(k*" ",i*"*   "," \n")
   k=k-2
  for i in range(3*n-2,2*n-2,-1):
   print(l*" ",i*"*   "," \n")
   l=l+2
  for i in range(2*n,3*n-1,1):
   print(o*" ",i*"*   "," \n")
   o=o-2
  for i in range(n-1,0,-1):
   print(m*" ",i*"*   "," \n")
   m=m+2
```
