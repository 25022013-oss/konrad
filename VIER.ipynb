```python
#W4A2
from math import sqrt
n = float(input("Nhập n để kiểm tra xem n có phải số nguyên tố hay không: "))
while n <= 0 or n % 1 !=0:
  n = float(input("Nhập lại : "))
def checkSoNguyenTo(n):
  PosInt = int(n)
  cotmoc = sqrt(PosInt)//1
  if (PosInt%2==0 and PosInt!=2) or PosInt == 1:
    kq = "False"
  else:
    i = cotmoc
    while 1<=i<=cotmoc:# nếu số nguyên này ko chia hết cho các số nguyên chạy từ 1 đến căn bậc 2 của nó ---> chắc chắn là số nguyên tố
      if i == 1:
        kq = "True"
        break
      if PosInt % i != 0:
        i = i-1
      else :
        kq = "False"
        break
  return kq
print(checkSoNguyenTo(n))



```

```python
#W4A4
n = int(input("Nhập n :"))
def demchuso(n) :
 if n < 0:
   absvalue = n*(-1)
 if n >= 0:
   absvalue = n
 count = 0
 if absvalue == 0:
   count = 1
 else:
  while absvalue != 0 :
    thuong = absvalue / 10
    if thuong < 0.1 :
     break
    else :
     count = count + 1
     absvalue = thuong
 return count
print(f"Số chữ số : {demchuso(n)}")
```

```python
#W4A33
a = list(map(str,str(input("Nhập int : "))))
dsmoi = []
if len(a)<=3:
  for i in a:
    print(i,end="")
else:
 for i in range(len(a)+1-4,-1,-3):
   if i <3 :
     if i == 1:
       a[i] = "." + a[i]
       dsmoi.append(a[i+2])
       dsmoi.append(a[i+1])
       dsmoi.append(a[i])
       dsmoi.append(a[i-1])
       break
     if i == 2:
       a[i] = "." + a[i]
       dsmoi.append(a[i+2])
       dsmoi.append(a[i+1])
       dsmoi.append(a[i])
       dsmoi.append(a[i-1])
       dsmoi.append(a[i-2])
       break
     if i == 0:
       dsmoi.append(a[i+2])
       dsmoi.append(a[i+1])
       dsmoi.append(a[i])
       break
   else :
      a[i] = "." + a[i]
      dsmoi.append(a[i+2])
      dsmoi.append(a[i+1])
      dsmoi.append(a[i])
 for k in range(0,len(dsmoi),1):
   print(dsmoi[len(dsmoi)-(k+1)],end="")
```
