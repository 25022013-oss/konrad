```python
#W5A1
def Max_Of_Two(a,b):
  if a > b:
    max = a
  else:
    max = b
  return max
print(Max_Of_Two(33,55))
```

```python
#W5A4
def PERFECT(n):
  tong = 0
  for i in range(1,n):
    if n%i == 0:
      tong = tong + i
    if tong == n:
      kq = True
    else:
      kq = False
  return kq
print(PERFECT(24))
```

```python
#W5A6
def FACTORIAL(n):
  tich = 1
  if n == 0:
    tich = 1
  if n>0:
    for i in range(n,0,-1):
      tich = tich * i
  return tich
print(FACTORIAL(3))
```

```python
#W5A8
def HAMMINGDISTANCE(num1,num2):
   remain1 = int(num1)
   remain2 = int(num2)
   ds1 = []
   ds2 = []
   ds1final=[]
   ds2final=[]
   while remain1 >=1:
     du1 = remain1%2
     ds1.append(du1)
     remain1 = remain1//2
   while remain2 >=1:
     du2 = remain2%2
     ds2.append(du2)
     remain2 = remain2//2
   while len(ds1)>len(ds2):
     ds2.append(0)
   while len(ds2)>len(ds1):
     ds1.append(0)
   for i in range(len(ds1)-1,-1,-1):
     ds1final.append(ds1[i])
   for i in range(len(ds2)-1,-1,-1):
     ds2final.append(ds2[i])
   dodai = len(ds1final)
   count = 0
   for i in range(dodai):
     if ds1final[i] != ds2final[i]:
       count = count + 1
   return count
print(HAMMINGDISTANCE(21321835,87654321))







```
